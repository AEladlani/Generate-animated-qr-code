# Styled QR CODE

______

```python
import qrcode
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.moduledrawers import SquareModuleDrawer
from qrcode.image.styles.colormasks import SolidFillColorMask
```

```python
# The data you want to encode
data = "https://github.com/AEladlani"
# Create QR code instance
qr = qrcode.QRCode(version=1, error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10, border=2,)
qr.add_data(data)
qr.make(fit=True)
```

```python
# Generate styled QR (✅ fixed colors as RGB tuples)
img = qr.make_image(image_factory=StyledPilImage,module_drawer=SquareModuleDrawer(),
    color_mask=SolidFillColorMask(back_color=(155,155,155), front_color=(0,155,150))) 

img.save("styled_qrcode2.png")
print("done ✅")
```

```python
import matplotlib.pyplot as plt
import qrcode
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.moduledrawers import SquareModuleDrawer
from qrcode.image.styles.colormasks import SolidFillColorMask

def generate_qr(data, size, color1, color2):
    qr = qrcode.QRCode(version=1, error_correction=qrcode.constants.ERROR_CORRECT_H,
         box_size=size, border=2,)
    qr.add_data(data)
    qr.make(fit=True)
    img = qr.make_image(image_factory=StyledPilImage,module_drawer=SquareModuleDrawer(),
        color_mask=SolidFillColorMask(back_color=color1, front_color=color2)) 
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
    img.save("gen_qr.png")

#size should be between 5 and 15
#color1 and color2 should be given as tuples (maybe a set of choices is given)
generate_qr(data,15,(155,155,155),(0,155,150))
```

```python
import os
import segno
from urllib.request import urlopen
from pathlib import Path
from typing import Optional

def generate_animated_qr(data: str,scale: int = 5,
    background: Optional[str] = None, output: str = "qr_code.gif"):
    """    Generate a QR code with optional image or GIF background.
    Parameters
    ----------
    data : str        Data to encode in the QR code
    scale : int        QR scale size
    background : str | None URL or local path to an image/GIF
    output : str  Output file name (.gif or .png) """
    print("🔹 Generating QR code...")
    qr = segno.make_qr(data)
    if background:
        print("🔹 Loading background...")
        # URL background
        if background.startswith("http"):
            try:
                bg = urlopen(background)
            except Exception as e:
                raise ValueError(f"Failed to load URL: {e}")
        # Local file background
        else:
            bg_path = Path(background)
            if not bg_path.exists():
                raise FileNotFoundError("Background file not found")
            bg = bg_path.open("rb")
        qr.to_artistic( background=bg, target=output, scale=scale)
    else:
        # Simple QR without background
        qr.save(output, scale=scale)
    print(f"✅ Done! File saved as: {output}")
```

```python
import os
import segno
from urllib.request import urlopen
from pathlib import Path
from typing import Optional
def generate_animated_qr(data: str,scale: int = 5,
    background: Optional[str] = None, output: str = "qr_code.gif"):
    print("🔹 Generating QR code...")
    qr = segno.make_qr(data)
    if background:
        print("🔹 Loading background...")
        # URL background
        if background.startswith("http"):
            try:
                bg = urlopen(background)
            except Exception as e:
                raise ValueError(f"Failed to load URL: {e}")
        # Local file background
        else:
            bg_path = Path(background)
            if not bg_path.exists():
                raise FileNotFoundError("Background file not found")
            bg = bg_path.open("rb")
        qr.to_artistic( background=bg, target=output, scale=scale)
    else:
        # Simple QR without background
        qr.save(output, scale=scale)
    print(f"✅ Done! File saved as: {output}")
```

```python
def interactive_qr():
    data = input("📌 Enter data to encode: ").strip()
    scale_input = input("📏 Enter QR scale (default 5): ").strip()
    scale = int(scale_input) if scale_input.isdigit() else 5
    use_bg = input("🎨 Use background? (y/n): ").lower().strip()
    background = None
    is_gif = False
    if use_bg == "y":
        background = input("🖼 Enter image/GIF path or URL: ").strip()
        # Detect GIF
        if background.lower().endswith(".gif"):
            is_gif = True
    # Ask output name (without extension)
    output_name = input("💾 Output filename (without extension, default qr_code): ").strip()
    output_name = output_name or "qr_code"
    # Decide extension automatically
    extension = ".gif" if is_gif else ".png"
    output = output_name + extension
    print(f"📁 Output will be saved as: {output}")
    generate_animated_qr(data=data,scale=scale,
                         background=background, output=output)

interactive_qr()
```

```python
from PIL import Image
import os

# Load the GIF
gif_path = "animated_qrcode_local.gif"  # Replace with your actual GIF file
output_dir = "frames"
# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)
# Open the GIF file
with Image.open(gif_path) as im:
    frame = 0
    try:
        while True:
            # Save current frame as PNG
            im.save(f"{output_dir}/image-{frame}.png", format="PNG")
            frame += 1
            im.seek(frame)  # Move to next frame
    except EOFError:
        pass  # End of sequence
```
