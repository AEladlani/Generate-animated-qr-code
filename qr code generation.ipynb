```python
from IPython.display import Image, display
import segno
from urllib.request import urlopen
import matplotlib.pyplot as plt
```

# qr code without background

```python
#define data to put in qr code
data = "...."
qrcode = segno.make_qr(data)
# save code
output = "test.png"
qrcode.save(output, scale=9)
```

```python
qr = plt.imread("test.png")
plt.imshow(qr, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

# from link

```python
#define data to put in qr code
data = "...."
slts_qrcode = segno.make_qr(data)
#define a link of a gif image to add in background
url = urlopen("https://media3.giphy.com/media/v1.Y2lkPTc5MGI3NjExbTJzdmlnZTlubm54ZWFqcTF0dHNxMW5yanVxYzl5Y3E4Y2RsdzQ3OCZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/3o6ZtgvOkvR3KI6tz2/giphy.gif")
```

```python
#define a path to save the generated animated qr code
output = "animated_qrcode.gif"
#save the qr code
slts_qrcode.to_artistic(background=url,
    target=output,scale=10,)
```

```python
from IPython.display import Image, display
output = "animated_qrcode.gif"
display(Image(filename=output))
```

# gif from local folder

```python
#define data
data = "...."
#generate qr code
qrcode = segno.make_qr(data)
#animate qr code with gif from local folder
#define path to gif
path = "Rotating_earth.gif"
# define save path
output = "animated_qrcode_local.gif"
with open(path, "rb") as gif_file:
    slts_qrcode.to_artistic(background=gif_file,
        target=output,scale=15)
```

```python
display(Image(filename=output))
```
